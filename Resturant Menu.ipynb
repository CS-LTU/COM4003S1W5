### Restaurant Menu
This is an example of a restaurant menu.
The example includes a menu with different categories, and clicking on a category shows the items in that category.

### HTML (index.html)

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta http-equiv="X-UA-Compatible" content="IE=edge">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <link rel="stylesheet" href="styles.css">
    <title>Restaurant Menu</title>
</head>
<body>
    <h1>Restaurant Menu</h1>
    <nav id="menu-categories">
        <button onclick="showCategory('starters')">Starters</button>
        <button onclick="showCategory('main_courses')">Main Courses</button>
        <button onclick="showCategory('desserts')">Desserts</button>
        <button onclick="showCategory('drinks')">Drinks</button>
    </nav>
    <div id="menu-items">
        <div class="category" id="starters">
            <h2>Starters</h2>
            <p>Garlic Bread</p>
            <p>Bruschetta</p>
            <p>Caesar Salad</p>
        </div>
        <div class="category" id="main_courses">
            <h2>Main Courses</h2>
            <p>Spaghetti Bolognese</p>
            <p>Grilled Salmon</p>
            <p>Margherita Pizza</p>
        </div>
        <div class="category" id="desserts">
            <h2>Desserts</h2>
            <p>Tiramisu</p>
            <p>Chocolate Cake</p>
            <p>Ice Cream</p>
        </div>
        <div class="category" id="drinks">
            <h2>Drinks</h2>
            <p>Red Wine</p>
            <p>White Wine</p>
            <p>Beer</p>
        </div>
    </div>
    <script src="script.js"></script>
</body>
</html>
```



### CSS (styles.css)
```css
body {
    font-family: Arial, sans-serif;
}

h1 {
    text-align: center;
}

#menu-categories {
    text-align: center;
    margin-bottom: 20px;
}

button {
    margin: 5px;
    padding: 10px 20px;
    font-size: 16px;
    cursor: pointer;
}

.category {
    display: none;
    text-align: center;
}

.category h2 {
    font-size: 24px;
    margin-bottom: 10px;
}

.category p {
    font-size: 18px;
    margin: 5px 0;
}

.active {
    display: block;
}
```



### JavaScript (script.js)
```js
function showCategory(categoryId) {
    const categories = document.querySelectorAll('.category');
    categories.forEach(category => {
        category.classList.remove('active');
    });
    const activeCategory = document.getElementById(categoryId);
    activeCategory.classList.add('active');
}
```



In this example:

1. The HTML structure includes a navigation section with buttons for each menu category, and a content section with the menu items divided into categories.
2. The CSS styles provide a clean and simple look, hiding the menu items by default, and showing them when they are active.
3. The JavaScript function `showCategory` is used to show the menu items of the selected category when the respective button is clicked.

To use this code, create three separate files (index.html, styles.css, and script.js) and copy the respective codes into them. Make sure that the files are saved in the same directory so that they can link correctly. Open the HTML file in a web browser to view and interact with the restaurant menu.
